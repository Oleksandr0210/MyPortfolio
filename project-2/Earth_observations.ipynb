EN **Geophysical and demographical observations**

Based on the data collection and analysis, a program was created that displays geophysical observations in Zakarpattia and demographic observations in different parts of the world in the form of line and dot plots using the Pandas and Matplotlib libraries.

**Stack:** Pandas, Pandas, Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
deformation_23 = pd.read_excel("deformation_2023.xlsx")
deformation_23
```

```python
deformation_23['*0138 мкм (10-7)'].plot(kind='line', figsize=(8, 4), title='Earth crustal movements in Oaș deep fault zone', ylabel='*0138 mkm (10-7)')
plt.gca().spines[['top', 'right']].set_visible(False)
```
